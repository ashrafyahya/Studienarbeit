```python
import RPi.GPIO as GPIO
import time

# GPIO-Nummerierung verwenden
GPIO.setmode(GPIO.BCM)

# Pin-Nummer für den Buzzer
BUZZER_PIN = 18

# Richten Sie den Pin als Ausgang für PWM ein
GPIO.setup(BUZZER_PIN, GPIO.OUT)
pwm = GPIO.PWM(BUZZER_PIN, 1)  # Starte PWM mit 1 Hz, die Frequenz wird später angepasst

try:
    pwm.start(50)  # Start PWM mit 50% Duty Cycle
    while True:
        for freq in range(100, 2000, 100):
            pwm.ChangeFrequency(freq)
            print(f"Buzzer-Frequenz: {freq} Hz")
            time.sleep(0.5)

except KeyboardInterrupt:
    print("Beenden durch Benutzer")

finally:
    pwm.stop()
    GPIO.cleanup()
```
