# Einführung in Matplotlib

### 1. Was ist Matplotlib ?
Matplotlib ist eine weit verbreitete [Python-Bibliothek](https://geekflare.com/de/popular-python-libraries-modules/) zur Visualisierung von Daten. Es ermöglicht Benutzern, ansprechende und leicht interpretierbare Diagramme zu erstellen, die dabei helfen, komplexe Daten besser zu verstehen. [Matplotlib](https://databasecamp.de/python/matplotlib) ist besonders bei Wissenschaftlern, Ingenieuren und Datenanalysten beliebt, da es eine einfache Möglichkeit bietet, Daten zu visualisieren.

### 2. Vorteile von Matplotlib
[Matplotlib](https://databasecamp.de/python/matplotlib) hat viele Vorteile, hier sind einpaar davon:  
***Einfach zu erlernen:*** Matplotlib hat eine intuitive Syntax, die besonders für Anfänger leicht verständlich ist.  
***Vielseitigkeit:*** Von einfachen [Linienplots](https://statorials.org/de/linearer-graph/) bis hin zu komplexen [3D-Diagrammen](https://www.delftstack.com/de/howto/seaborn/seaborn-3d-plot/) unterstützt Matplotlib eine Vielzahl von Diagrammtypen.  
***Anpassbarkeit:*** Benutzer können fast jedes Detail eines Diagramms anpassen, von Farben über Achsenbeschriftungen bis hin zu Linienbreiten.  
***Kompatibilität:*** Matplotlib funktioniert hervorragend mit anderen Python-Bibliotheken wie [NumPy](https://databasecamp.de/python/numpy), [Pandas](https://databasecamp.de/python/pandas-grundlagen-1) und [Seaborn](https://databasecamp.de/python/seaborn), was die Datenvisualisierung für komplexe Analysen erleichtert.  
***Vielseitiges Ausgabeformat:*** Plots können in verschiedenen Formaten gespeichert werden, was nützlich für Publikationen oder Präsentationen ist.

***3.3 erster Schritt in Matplotlib:*** Es ist ein einfaches Beispiel für Darstellung eines Plots in Matplotlib:

`````python
### Imporieren Sie die Bibliotheken
import matplotlib.pyplot as plt

# Einfache Linie zeichnen
plt.plot([2, 4, 6, 8], [2, 4, 8, 12])
plt.title("Ein einfacher Plot")
plt.show()
`````

### 3.4 Erklärung des Codes
````import matplotlib.pyplot as plt````  : Importiert die [pyplot](https://matplotlib.org/stable/tutorials/pyplot.html)-Funktionen von Matplotlib, um einfache Plot-Befehle zu verwenden.  
````plt.plot()````  : Erstellt einen Linienplot basierend auf den angegebenen x- und y-Daten.  
````plt.title()````  : Fügt einen Titel zum Diagramm hinzu.

### 4.Funktionen von Matplotlib
Die typischen Funktionen von Matplotlib können wie folgt gelistet werden:

| Funktion                         | Beschreibung                                                                                                   |
|----------------------------------|---------------------------------------------------------------------------------------------------------------|
| **Vielfältige Diagrammtypen**    | Matplotlib unterstützt eine Vielzahl von Diagrammtypen, darunter Linien-, Balken-, Streu- und Kreisdiagramme sowie 3D-Visualisierungen. |
| **Anpassung**                    | Die meisten Aspekte eines Diagramms können detailliert angepasst werden (z.B. Titel, Achsen, Farben).         |
| **Mehrere Plots**                | Mit `subplots()` können mehrere Diagramme auf einer Abbildung erstellt werden, was es einfach macht, verschiedene Visualisierungen nebeneinander darzustellen. |
| **Interaktive Funktionen**        | In Kombination mit Jupyter Notebooks bietet Matplotlib interaktive Diagramme, bei denen Nutzer zoomen und Datenpunkte hervorheben können. |


### 6. Arten von Plots in Matplotlib

| Plot-Typ           | Beschreibung                                                    | Funktion                            |
|--------------------|----------------------------------------------------------------|-------------------------------------|
| **Linienplot**     | für zeitliche Trends und kontinuierliche Daten.               | `plot()`                            |
| **Streudiagramm**  | für die Visualisierung von Datenpunkten und deren Verteilung. | `scatter()`                         |
| **Balkendiagramm** | für den Vergleich von kategorischen Daten.                    | `bar()`                             |
| **Histogramm**     | für die Darstellung von Häufigkeitsverteilungen.               | `hist()`                            |
| **Kuchendiagramm** | für prozentuale Anteile an einem Ganzen.                      | `pie()`                             |
| **Boxplot**        | für die Darstellung von Verteilungen und Ausreißern.          | `boxplot()`                         |
| **Flächendiagramm**| für die Darstellung kumulativer Werte.                         | `fill_between()`                   |
| **Dichteplot**     | für die Visualisierung der Wahrscheinlichkeitsdichte.         | `hist()` mit der Option `density=True` |
| **Heatmap**        | für die Visualisierung von Matrizen und Korrelationen.        | `imshow()`                          |
| **3D-Plot**        | für dreidimensionale Visualisierungen.                         | `plot_surface()` aus dem Modul `mpl_toolkits.mplot3d` |


### 6.1 Beispiele zu Plots
hier werden einpaar Beispiele zu manschen der Plots gezeigt.

#### 6.1.1 Beispiel zum Balkendiagramm

`````python
plt.bar(['A', 'B', 'C'], [10, 20, 30])
plt.title("Balkendiagramm")
plt.show()
`````

#### 6.1.2 Beispiel zum Streudiagramm

`````python
plt.scatter([1, 2, 3, 4], [5, 10, 15, 20])
plt.title("Streudiagramm")
plt.show()
`````

#### 6.1.3 Beispiel zum Histodiagramm

`````python
import numpy as np
data = np.random.randn(100)
plt.hist(data, bins=10)
plt.title("Histogramm")
plt.show()
`````

### 7. Matplotlib im Vergleich zu anderen Bibliotheken
[***Seaborn:***](https://databasecamp.de/python/seaborn) Baut auf Matplotlib auf, bietet jedoch eine einfachere API und speziellere Diagramme für statistische Daten. Nutzen Seaborn, wenn Sie besonders ästhetische und komplexe statistische Plots erstellen möchten.

### 8. häufige Probleme und deren Lösungen
beim Erstellen von Diagramme, können Sie möglicherweise folgende Fehler sehen:  
***Kein Plot wird angezeigt:*** In Jupyter Notebooks sollen Sie ```%matplotlib inline``` hinzufügen, um Diagramme korrekt anzuzeigen.  
***Diagramme sind zu klein oder unleserlich:*** Verwenden Sie ````figsize```` in ````plt.figure()````, um die Größe der Abbildung anzupassen:

`````python
### Beispiel, um Diagramme groß und leserlich zu machen
plt.figure(figsize=(10, 6))
`````
